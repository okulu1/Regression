```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df= pd.read_csv('energydata_complete.csv')

df=df.drop(['date','lights'],axis=1)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
reg_min = MinMaxScaler()
normalized_df = pd.DataFrame(reg_min.fit_transform(df),columns=df.columns)
normalized_df
```

```python
target1 = normalized_df['Appliances']
features= normalized_df.drop(['Appliances'],axis=1)
features
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features,target1,
test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(x_train,y_train)
```

```python
y_hat = linear_model.predict(x_test)
y_hat
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_hat)
round(mae, 2)
```

```python
import numpy as np
rss = np.sum(np.square(y_test - y_hat))
round(rss, 2)
```

```python
#Root Mean Square Error (RMSE)
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_hat))
round(rmse, 3)
```
